```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
x=range(1,200)
y=[0.015*i for i in x]
z=[-0.011*i+2 for i in x]
a=[-0.011*i+2.5 for i in x]
fig,ax=plt.subplots()
ax.plot(x,y,label='供给曲线')
ax.plot(x,z,label='需求曲线')
ax.plot(x,a,label='移动后的需求曲线')
ax.legend(loc='upper left')  # Add a legend.
ax.set_xlabel('冰淇淋的数量(个)')
ax.set_ylabel('冰淇淋的价格(元)')
```
