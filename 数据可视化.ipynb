```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,5,0.1)
y=np.sin(x)
plt.plot(x,y,'ro')
plt.axis([1,3,0,1])#The :func:~matplotlib.pyplot.axis command in the example above takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes.
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))
plt.style.use(['tableau-colorblind10','ggplot'])

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

*available方法用来查看pyplot里的样式种类*

```python
plt.style.available
```

**用如下的循环，可以查看样式**

```python
styles=plt.style.available
for a in styles:
    x=np.arange(0,2*np.pi,0.1)
    y=np.sin(x)
    plt.plot(x,y)
    plt.style.use(a)
    plt.show()
```
