```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
html=urlopen('http://feng.li/python/')
```

```python
bs=BeautifulSoup(html.read(),'html.parser')
```

```python
bs
```

```python
namelist=bs.find_all('div',{'class':'entry-content'})
for name in namelist:
    print(name.get_text())
```
